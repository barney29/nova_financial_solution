```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
import talib
```

```python
#Loding the data for google 
df_google = yf.download('GOOG')
```

## Moving Average

### 1. Simple moving average

```python
df_google['SMA_10'] = df_google['Adj Close'].rolling(window=10, min_periods=1).mean()
```

```python
df_google[['Adj Close', 'SMA_10']].plot(figsize=(15,6), title="Simple moving average window size 10 for Adj Close")
```

```python
#Not clear to see so extending window frame to 50
df_google['SMA_50'] = df_google['Open'].rolling(window=50, min_periods=1).mean()
```

```python
#more smooth than window size 10
df_google[['Open', 'SMA_50']].plot(figsize=(15,6), title="Simple moving average for Open")
```

### 2. Exponential Weighted Moving Average

```python
df_google['EWMA'] = df_google['Open'].ewm(alpha=0.1, adjust=False).mean()
```

```python
df_google[['Open', 'EWMA']].plot(figsize=(15,6))
```

## Relative Strength Indicator

```python
talib.RSI(df_google['Open'], timeperiod=14).plot(figsize=(15,6), title="Relative Strength Indicator for Open")
```

```python
df_google['MACD'], df_google['Signal_line'], _ = talib.MACD(df_google['Adj Close'].values, fastperiod=12, slowperiod=26, signalperiod=9)
```

```python
df_google[['Adj Close', 'MACD', 'Signal_line']].plot(figsize=(15,6))
```

```python
# ploting all the data for MACD and the Signal_line

#drawing small sample

df_google.loc['2023-1-1':'2024-1-1'][['MACD', 'Signal_line']].plot(figsize=(15,6))


```
